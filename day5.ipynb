```python
def getRow(binary_seat_id):
    row_code = binary_seat_id[:7]
    row_number = 0
    step = 6
    for char in row_code:
        if (char=="B"):
            row_number+=(2**step)
        step-=1     
    return row_number
    
# Validate method
getRow("FBFBBFFRLR")
```

```python
def getSeat(binary_seat_id):
    seat_code = binary_seat_id[7:10]
    seat_number = 0
    step = 2
    for char in seat_code:
        if (char=="R"):
            seat_number+=(2**step)
        step-=1     
    return seat_number

# Validate method
getSeat("FBFBBFFRLR")
```

```python
def calculateSeatId(binary_seat_id):
    return getRow(binary_seat_id) * 8 + getSeat(binary_seat_id)

# Validate method
calculateSeatId("FBFBBFFRLR")
```

```python
seat_ids = [calculateSeatId(binary_seat_id) for binary_seat_id in input_array]
# Answer to part 1
print(max(seat_ids))
```
