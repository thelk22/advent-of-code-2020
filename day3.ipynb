```python
import numpy as np

f = open("day3_input.txt", "r")

# Read in the date from file
input_array = np.array(f.read().split("\n"))

# Remove the whitespace entry
input_array = input_array[:-1]

# Verify output array by eye
print(input_array)
```

```python
# Part 1

rows = len(input_array)
cols = len(input_array[0])
# print(rows, cols)

tree_count = 0
j = 0
for i in range(rows):
    char = input_array[i][j]
    if (char=='#'):
        tree_count+=1
#     Check row, column, character
#     print(i,j,char)
    j = (j + 3) % cols
    
print(tree_count)
    
    
    
```

```python
# Part 2
# Use a method...

def find_number_of_trees_for_slope(input_array, steps_right, steps_down):
    rows = len(input_array)
    cols = len(input_array[0])
    # print(rows, cols)

    tree_count = 0
    j = 0
    for i in range(0,rows,steps_down):
        char = input_array[i][j]
        if (char=='#'):
            tree_count+=1
    #     Check row, column, character
    #     print(i,j,char)
        j = (j + steps_right) % cols

    return tree_count

result_1 = find_number_of_trees_for_slope(input_array, 1,1)
result_2 = find_number_of_trees_for_slope(input_array, 3,1)
result_3 = find_number_of_trees_for_slope(input_array, 5,1)
result_4 = find_number_of_trees_for_slope(input_array, 7,1)
result_5 = find_number_of_trees_for_slope(input_array, 1,2)

answer = result_1 * result_2 * result_3 * result_4 * result_5

print(answer)



```
