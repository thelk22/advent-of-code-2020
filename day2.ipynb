```python
f = open("day2_input.txt", "r")

# Read in the date from file
input_array = f.read().split("\n")

# Remove the whitespace entry
input_array = input_array[:-1]

# Format output array for running through algorithm
output_array = []
for i in input_array:
    items = i.split(" ");
    minimum = int(items[0].split("-")[0]);
    maximum = int(items[0].split("-")[1]);
    character = items[1][:1];
    password = items[2];
    output_array.append([character, minimum, maximum, password]);

# Verify output array by eye
print(output_array)
```

```python
def total_valid_passwords_1(array):
    valid_passwords = 0
    for a in array:
        character = a[0]
        min_allowed = a[1]
        max_allowed = a[2]
        password = a[3]
        if (min_allowed <= password.count(character) <= max_allowed):
            valid_passwords+=1;
    return valid_passwords

print(total_valid_passwords_1(output_array))
```

```python
def total_valid_passwords_2(array):
    valid_passwords = 0
    for a in array:
        character = a[0]
        position_1 = a[1]
        position_2 = a[2]
        password = a[3]
        
        character_in_position_1 = character==password[position_1-1]
        character_in_position_2 = character==password[position_2-1]
        
        if (character_in_position_1 + character_in_position_2 == 1):
            valid_passwords+=1;
            
            
    return valid_passwords

print(total_valid_passwords_2(output_array))
```
